## Information about dataset attributes :-

* **Product_id** : A unique identifier for each product.
* **Price** : The price of the product.
* **Sale** : Indicates if the product is on sale or if there are any discounts.
* **Weight** : The weight of the product.
* **Resolution** : The resolution of the product, possibly related to the display or camera.
* **PPI (Pixels Per Inch)** : A measure of the pixel density in a display.
* **CPU Core** : The number of CPU cores in the product's processor.
* **CPU Freq (CPU Frequency)** : The clock speed or frequency of the CPU.
* **Internal Mem (Internal Memory)** : The internal storage capacity of the product.
* **RAM** : The random access memory (RAM) capacity of the product.
* **RearCam (Rear Camera)** : Specification related to the resolution or features of the rear camera.
* **Front_Cam (Front Camera)** : Specification related to the resolution or features of the front-facing camera.
* **Battery** : The capacity or power of the product's battery.
* **Thickness** : The thickness or depth of the product

# MODEL

```python
data.sample(5)
```

```python
from sklearn.model_selection import train_test_split  

X = data.drop('Price' ,axis=1)  
y = data['Price']
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0  )  
```

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
```

# LogisticRegression

```python
from sklearn.linear_model import LogisticRegression  
classifier1= LogisticRegression(random_state = 0 )  
classifier1.fit(X_train, y_train)     
```

```python
y_pred1=classifier1.predict(X_test)  

from sklearn.metrics import accuracy_score, confusion_matrix  
a =accuracy_score(y_test, y_pred1)   
print(a)
```

```python
cm1 = confusion_matrix(y_test, y_pred1)
print(cm1)
```

# Support Vector Machine (SVM)

```python
from sklearn.svm import SVC
classifier2 = SVC(kernel = 'linear', random_state = 0)
classifier2.fit(X_train, y_train)
```

```python
y_pred2 = classifier2.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
b=accuracy_score(y_test, y_pred2) 
print(b)
```

```python
cm2 = confusion_matrix(y_test, y_pred2)
print(cm2)
```

# Decision Tree Classification

```python
from sklearn.tree import DecisionTreeClassifier

classifier3 = DecisionTreeClassifier( criterion = 'entropy', random_state = 0 ,max_depth=5)
classifier3.fit(X_train, y_train)
```

```python
y_pred3 = classifier3.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
c=accuracy_score(y_test, y_pred3) 
print(c)
```

```python
cm3 = confusion_matrix(y_test, y_pred3)
print(cm3)
```

# K-Nearest Neighbors (K-NN)

```python
from sklearn.neighbors import KNeighborsClassifier

classifier4 = KNeighborsClassifier(n_neighbors = 5, metric = 'minkowski', p = 2)
classifier4.fit(X_train, y_train)
```

```python
y_pred4 = classifier4.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
d=accuracy_score(y_test, y_pred4) 
print(d)
```

```python
cm4 = confusion_matrix(y_test, y_pred4)
print(cm4)
```

# Naive Bayes

```python
from sklearn.naive_bayes import GaussianNB

classifier5 = GaussianNB()
classifier5.fit(X_train, y_train)
```

```python
y_pred5 = classifier5.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
e=accuracy_score(y_test, y_pred5) 
print(e)
```

```python
cm5 = confusion_matrix(y_test, y_pred5)
```

# Random Forest Classification

```python
from sklearn.ensemble import RandomForestClassifier

classifier6 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier6.fit(X_train, y_train)
```

```python
y_pred6 = classifier6.predict(X_test)
from sklearn.metrics import confusion_matrix, accuracy_score
f=accuracy_score(y_test, y_pred6) 
print(f)
```

```python
cm6 = confusion_matrix(y_test, y_pred6)
```

```python
col=['LogisticRegression','Support Vector Machine','Decision Tree Classification','K-Nearest Neighbors','Naive Bayes','Random Forest Classification']
result=[a,b,c,d,e,f]
```
